## Data Preprocessing

```python
house_df = house_df.drop(['id','date','zipcode'],axis=1)
house_df.head()
```

```python
house_df['age'] = 2018 - house_df['yr_built']
house_df = house_df.drop('yr_built',axis=1)
house_df.head()
```

```python
house_df['renovated'] = house_df['yr_renovated'].apply(lambda x:1 if x>0 else 0)
house_df=house_df.drop('yr_renovated',axis=1)
house_df.head()
```

```python
house_df.to_csv("./datasets/house_data_preprocessed.csv",index=False)
```

## Modelling

```python
from sklearn.model_selection import train_test_split
X = house_df.drop('price',axis=1)
Y = house_df['price']

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg_model = LinearRegression().fit(x_train,y_train)
```

```python
print("Training score : ", reg_model.score(x_train,y_train))
#regression model uses r-square to evaluate the model. Basically r-square gives the value of how of the 
# underlying variance is captured by our model
```

```python
predictors = x_train.columns
coef = pd.Series(reg_model.coef_,predictors).sort_values()
print(coef)
```

```python
y_pred = reg_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print("Testing score :" , r2_score(y_test,y_pred))
```

```python
import statsmodels.api as sm
x_train_with_const = sm.add_constant(x_train)
model = sm.OLS(y_train,x_train_with_const)
#OLS = Ordinary Least Square
```

```python
result = model.fit()
```

```python
print(result.summary())
```

R-squared value : This value shows how well our model captured the underlying data.  
Adjusted R-Square value : this is a modification of R-square. This value will fall if we have features in our dataset that does not add value. 
F-statistic and Prob shows the significance of our result. High value of f-statistic with low value of probability tell that our test is significant.  
P-value shows that probability that our score was obtained by chance. Low P-value is good
