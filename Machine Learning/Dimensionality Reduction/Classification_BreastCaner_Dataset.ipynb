```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

```python
print(cancer_dataset.DESCR)
```

```python
cancer_df = pd.DataFrame(data=cancer_dataset.data,columns = cancer_dataset.feature_names)
cancer_df['class'] = cancer_dataset.target
```

```python
cancer_df.sample(4)
```

This dataset has lots of features. We can check some of the features and check their scatterplot

```python
fig,ax = plt.subplots(figsize=(10,8))
plt.scatter(cancer_df['radius error'],cancer_df['mean radius'])

plt.xlabel('radius error')
plt.ylabel('mean radius')
```

```python
cancer_df.corr()
```

```python
fig,ax = plt.subplots(figsize=(10,8))
sns.heatmap(cancer_df.corr())
```

```python
from sklearn.model_selection import train_test_split
X = cancer_df.drop('class',axis=1)
Y = cancer_df['class']

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(solver = 'liblinear')
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
pd.crosstab(y_pred,y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

print("accuracy score ",accuracy_score(y_test,y_pred))
print("precision score ",precision_score(y_test,y_pred))
print("recall score ",precision_score(y_test,y_pred))
```

accuracy score of 95% means that 95% of the instances were correctly classified  
precision score of 94% means that out of all the instances that our model said the intance was malignant,it was correct 94% of the time  
recall score of 94% means that out of all the malignant instances, our model was able to correctly idenity 94% of them
